3. How many orders were made within the date period

```python
%sql select count(order_id) as 'No. of orders' from order_details;
```

4. How many items were ordered within the date range

```python
%sql select count(distinct (order_id)) as 'items_ordered' from order_details;
```

5. Which orders with the most no of items 

```python
%%sql select order_id, count(item_id) as 'items_ordered' from order_details
        group by order_id
        order by items_ordered
        DESC limit 10;
        
```

6. How many orders had items more than 12

```python
%%sql select count(*) as 'no_orders' from 
        (select order_id, count(item_id) as 'items_ordered' from order_details
        group by order_id
        having items_ordered > 12) as num_orders;
```

LET'S ANALYZE CUSTOMER BEHAVIOUR.
1. JOIN THE TWO TABLES

```python
%%sql select * from order_details od
        join menu_items mi
        on od.item_id = mi.menu_item_id
        limit 10;
```

2. What were the least and most ordered items? What categories were they in? 

```python
%%sql select item_name, category, count(order_details_id) as 'no_of_orders'
        from order_details od
        join menu_items mi
        on od.item_id = mi.menu_item_id
        group by item_name, category
        order by no_of_orders desc limit 1;
```

```python
%%sql select item_name, category, count(order_details_id) as 'no_of_orders'
        from order_details od
        join menu_items mi
        on od.item_id = mi.menu_item_id
        group by item_name, category
        order by no_of_orders asc limit 1;
```

3. What were the top 5 orders that spent most money?

```python
%%sql select order_id, sum(price) as 'money_on_orders'
        from order_details od
        join menu_items mi
        on od.item_id = mi.menu_item_id
        group by order_id
        order by money_on_orders desc limit 5;
```

4. Let's view details of top 5 spent orders

```python
%%sql select *
        from order_details od
        join menu_items mi
        on od.item_id = mi.menu_item_id
        where order_id in (440,2075,1957,330,2675);
```

What categories did the most items ordered come from

```python
%%sql select order_id, category, count(item_id) as 'no_items' 
        from order_details od
        join menu_items mi
        on od.item_id = mi.menu_item_id
        where order_id in (440,2075,1957,330,2675)
        group by order_id, category;
```
